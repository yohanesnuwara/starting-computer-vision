```python
import numpy as np
import matplotlib.pyplot as plt
import gdal
import cv2
from google.colab.patches import cv2_imshow
```

```python
RGB_img = cv2.imread('/content/fruits.jpeg')

cv2_imshow(RGB_img)
```

```python
RGB_img
```

```python
plt.figure(figsize=(5,7))
for i in range(3):
  plt.subplot(2,2,i+1)
  plt.imshow(RGB_img[:,:,i], aspect='auto')
  plt.title(f'Channel {i+1}')

plt.tight_layout()
plt.show()
```

```python
pear = RGB_img[500:600,400:,:]
avocado = RGB_img[100:200,150:200,:]
orange = RGB_img[400:500,25:75,:]
apple = RGB_img[150:250,250:300,:]
cv2_imshow(pear)
cv2_imshow(avocado)
cv2_imshow(orange)
cv2_imshow(apple)
```

```python
mat = [pear, avocado, orange, apple]
mat_type = ['Pear', 'Avocado', 'Orange', 'Apple']

plt.figure(figsize=(10,3))
for i in range(3):
  plt.subplot(1,3,i+1)
  plt.title(f'Channel {i+1}')
  for j in range(4):
    m = mat[j]
    ch = m[:,:,i]
    plt.hist(ch.flatten(), label=mat_type[j], alpha=.5)
  plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Orange/lemon
mask = cv2.inRange(RGB_img, (0, 150, 225), (50, 250, 250))

imask = mask>0
green = np.zeros_like(RGB_img, np.uint8)
green[imask] = RGB_img[imask]

plt.imshow(mask)
```
